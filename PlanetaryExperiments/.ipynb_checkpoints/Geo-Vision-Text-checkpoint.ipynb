### Import HuggingFace model

```python
#!pip install huggingface_hub
#!pip install open-clip-torch

from huggingface_hub import hf_hub_download
import torch, open_clip

# Download model and setup
model_name = 'RN50'  # 'RN50', 'ViT-B-32', or 'ViT-L-14'
checkpoint_path = hf_hub_download("chendelong/RemoteCLIP", f"RemoteCLIP-{model_name}.pt", cache_dir='checkpoints')
model, _, preprocess = open_clip.create_model_and_transforms(model_name)
model.load_state_dict(torch.load(checkpoint_path, map_location="cpu"))
model.eval()
```

### Define the folder, chip size, and query

```python
from PIL import Image
import rasterio
from rasterio.plot import reshape_as_image
import numpy as np
import os, glob

image_folder = "Assets"
chip_size = 200  # size of the chips in pixels
query = "give me the top 5 images that show a creek besides a house"
top_n = 5

# Tokenize the query
text = open_clip.tokenize([query])
text_features = model.encode_text(text)
```

### Process and chip the image

```python
import rasterio
from rasterio.warp import transform_bounds
from rasterio.plot import reshape_as_image

def process_image(filepath, chip_size):
    with rasterio.open(filepath) as src:
        # Assuming the source images are not in WGS84, get the CRS of the source
        src_crs = src.crs
        for x in range(0, src.width, chip_size):
            for y in range(0, src.height, chip_size):
                window = rasterio.windows.Window(x, y, chip_size, chip_size)
                if src.width - x < chip_size or src.height - y < chip_size:
                    continue
                chip = src.read(window=window)
                chip_image = reshape_as_image(chip)
                if np.count_nonzero(chip_image) > 0:
                    # Get the geographic bounds of the window and transform to WGS84
                    bounds = src.window_bounds(window)
                    wgs84_bounds = transform_bounds(src_crs, 'EPSG:4326', *bounds)
                    yield chip_image, wgs84_bounds
```

### Process images and subdivide into chips

```python
import glob
from tqdm import tqdm

filepaths = list(glob.glob(f'{image_folder}/*.tif'))
total_chips = sum(1 for filepath in filepaths for _ in process_image(filepath, chip_size))

chip_scores = []
progress_bar = tqdm(total=total_chips, desc='Processing chips')

for filepath in filepaths:
    for chip_image, bounds in process_image(filepath, chip_size):
        chip_preprocessed = preprocess(Image.fromarray(chip_image)).unsqueeze(0)
        chip_features = model.encode_image(chip_preprocessed)
        similarity = (chip_features @ text_features.T).item()
        chip_scores.append((filepath, similarity, chip_image, bounds))
        progress_bar.update(1)


progress_bar.close()
```

### Sort chips by similarity and select top N

```python
top_chips = sorted(chip_scores, key=lambda x: x[1], reverse=True)[:top_n]

# Display or process top chips
for filepath, score, chip_image, bounds in top_chips:
    print(f"Chip from {filepath} with score {score}, bounds: {bounds}")
    display(Image.fromarray(chip_image))
```

```python
from shapely.geometry import box
import geopandas as gpd
import leafmap.foliumap as leafmap

# Create a GeoDataFrame from the bounds
gdf = gpd.GeoDataFrame(geometry=[box(*bounds) for _, _, _, bounds in top_chips])

# Initialize the map with Google Satellite as the basemap
m = leafmap.Map(basemap='GOOGLE_SATELLITE')

# Add the GeoDataFrame as a layer
m.add_gdf(gdf, layer_name='Top Chips', style={"color": "blue"})

# Zoom to the extent of the GeoDataFrame
m.zoom_to_gdf(gdf)

m
```

### Save Chips in case they are needed later

```python
import os
from PIL import Image

# Directory to save the chips
chip_folder = 'Assets/Chips'
os.makedirs(chip_folder, exist_ok=True)

filepaths = list(glob.glob(f'{image_folder}/*.tif'))
progress_bar = tqdm(total=total_chips, desc='Processing and saving chips')

for filepath in filepaths:
    for i, chip_image in enumerate(process_image(filepath, chip_size)):
        # Convert the numpy array to an image
        chip_image_pil = Image.fromarray(chip_image)
        
        # Construct a filename for the chip
        base_filename = os.path.splitext(os.path.basename(filepath))[0]
        chip_filename = f"{base_filename}_chip_{i}.png"
        chip_path = os.path.join(chip_folder, chip_filename)
        
        # Save the chip image
        chip_image_pil.save(chip_path)
        
        # Update progress bar
        progress_bar.update(1)

progress_bar.close()
```

### Pre-process saved chips
You'll notice it takes longer to read from disk

```python
import glob
from tqdm import tqdm

# Assuming 'preprocess', 'model', and 'text_features' are already defined
chip_folder = 'Assets/Chips'
chip_files = glob.glob(f'{chip_folder}/*.png')

chip_scores = []
progress_bar = tqdm(total=len(chip_files), desc='Processing chips')

for chip_file in chip_files:
    with Image.open(chip_file) as chip_image:
        chip_preprocessed = preprocess(chip_image).unsqueeze(0)
        chip_features = model.encode_image(chip_preprocessed)
        similarity = (chip_features @ text_features.T).item()
    chip_scores.append((chip_file, similarity))
    progress_bar.update(1)


progress_bar.close()
```

#### Select the top chips

```python
top_chips = sorted(chip_scores, key=lambda x: x[1], reverse=True)[:5]

for filepath, score in top_chips:
    print(f"Chip from {filepath} with score {score}")
    with Image.open(filepath) as img:
        display(img)
```

### Generate GeoJson of the top chips

```python
import rasterio
from rasterio.features import bounds
import geopandas as gpd
from shapely.geometry import box

def create_geojson(top_chips, chip_size):
    features = []
    for filepath, score in top_chips:
        with rasterio.open(filepath) as src:
            chip_index = int(os.path.splitext(os.path.basename(filepath))[0].split('_')[-1])
            rows = src.height // chip_size
            x = (chip_index % rows) * chip_size
            y = (chip_index // rows) * chip_size
            window = rasterio.windows.Window(x, y, chip_size, chip_size)
            geom = box(*src.window_bounds(window))
            features.append({'type': 'Feature', 'properties': {'score': score}, 'geometry': geom.__geo_interface__})

    gdf = gpd.GeoDataFrame.from_features(features, crs=src.crs)
    gdf.to_file('top_chips.geojson', driver='GeoJSON')

# Assuming top_chips is a list of filepaths for the top 5 images
top_chips_with_score = sorted(chip_scores, key=lambda x: x[1], reverse=True)[:5]
# Assuming top_chips_with_score contains the top 5 chips with their scores
create_geojson(top_chips_with_score, chip_size=200)
```
